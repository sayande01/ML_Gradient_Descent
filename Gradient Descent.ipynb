```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the quadratic function (example cost function)
def quadratic_function(x):
    return x**2
```

```python
# Define the derivative of the quadratic function (gradient)
def gradient_quadratic(x):
    return 2*x
```

```python
# Gradient Descent function
def gradient_descent(learning_rate, epochs):
    # Initialize parameter (starting point)
    x = 4.0
    # Lists to store parameter and loss values for visualization
    parameter_values = [x]
    loss_values = [quadratic_function(x)]   
```

```python
# Gradient Descent iterations
    for epoch in range(epochs):
        # Compute gradient
        grad = gradient_quadratic(x)
        # Update parameter using gradient descent
        x = x - learning_rate * grad
        # Compute loss
        loss = quadratic_function(x)
        # Append parameter and loss values
        parameter_values.append(x)
        loss_values.append(loss)
    
    return parameter_values, loss_values
```

```python
# Define hyperparameters
learning_rate = 0.1
epochs = 10
```

```python
# Run Gradient Descent
parameters, losses = gradient_descent(learning_rate, epochs)
```

```python
# Plot the quadratic function
x_vals = np.linspace(-5, 5, 100)
y_vals = quadratic_function(x_vals)
plt.plot(x_vals, y_vals, label='Quadratic Function')

# Plot the gradient descent steps
plt.scatter(parameters, losses, color='red', label='Gradient Descent Steps')
plt.xlabel('Parameter (x)')
plt.ylabel('Loss')
plt.title('Gradient Descent Optimization')
plt.legend()
plt.grid(True)
plt.show()
```
